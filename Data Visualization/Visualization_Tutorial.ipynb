# ðŸŒ¸ Data Visualization with the Iris Dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from scipy import stats

# Load the Iris dataset
iris = sns.load_dataset('iris')
sns.set_theme()
```

## ðŸ“Œ Scatter Plots

> **Use when**: Use this when you want to explore the relationship between two continuous variables.

### Simple Scatter Plot

```python
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width')
plt.show()
```

> **Use when**: Use this to assess the strength and direction of a linear relationship between two variables.

### Scatter Plot with Regression Line

```python
sns.regplot(data=iris, x='sepal_length', y='sepal_width', ci=None)
plt.show()
```

> **Use when**: Use this to visualize non-linear trends in a scatter plot.

### Fitted Line (Lowess)

```python
sns.lmplot(data=iris, x='sepal_length', y='sepal_width', lowess=True)
plt.show()
```

> **Use when**: Use this to compare trends across categories within your dataset.

### Grouped Scatter Plot with Regression Lines

```python
sns.lmplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
plt.show()
```

> **Use when**: Use this to explore all pairwise relationships in a multivariate dataset.

### Pairplot (Scatterplot Matrix)

```python
sns.pairplot(iris, hue='species')
plt.show()
```

## ðŸ“Œ Exploring Distributions

> **Use when**: Use this to understand the distribution of a single numeric variable.

### Simple Histogram

```python
sns.histplot(data=iris, x='petal_length')
plt.show()
```

> **Use when**: Use this to see how the choice of bin width affects the perceived distribution shape.

### Histogram with Varying Bin Widths

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
sns.histplot(data=iris, x='petal_length', bins=10)
plt.title('10 Bins')

plt.subplot(1, 2, 2)
sns.histplot(data=iris, x='petal_length', bins=30)
plt.title('30 Bins')
plt.tight_layout()

plt.show()
```

> **Use when**: Use this to compare distributions of a numeric variable across different categories.

### Grouped Histogram

```python
sns.histplot(data=iris, x='petal_length', hue='species', element='step', stat='density', common_norm=False)
plt.show()
```

> **Use when**: Use this to show cumulative distribution across categories.

### Stacked Histogram

```python
sns.histplot(data=iris, x='petal_length', hue='species', multiple='stack')
plt.show()
```

> **Use when**: Use this to display a smooth estimate of the distribution for continuous data.

### KDE Plot

```python
sns.kdeplot(data=iris, x='petal_length', hue='species', fill=True)
plt.show()
```

> **Use when**: Use this to visualize individual data points and understand the data density without overlapping points.

### Swarm Plot

```python
sns.swarmplot(data=iris, x='species', y='petal_length')
plt.show()
```

## ðŸ“Œ Bar Plots

> **Use when**: Use this to compare the mean or total value across two categories.

### Simple Bar Plot

```python
data = iris[iris['species'].isin(['setosa', 'versicolor'])]
mean_values = data.groupby('species')['sepal_length'].mean().reset_index()
sns.barplot(data=mean_values, x='species', y='sepal_length')

plt.show()
```

> **Use when**: Use this to compare values across two categorical variables.

### Grouped Bar Plot

```python
df = iris.melt(id_vars='species', value_vars=['sepal_length', 'sepal_width'])
mean_df = df.groupby(['species', 'variable']).mean().reset_index()
sns.barplot(data=mean_df, x='variable', y='value', hue='species')

plt.show()
```

> **Use when**: Use this when you want to indicate the uncertainty of a value estimate.

### Bar Plot with Confidence Intervals

```python
sns.barplot(data=iris, x='species', y='petal_length', ci=95)
plt.show()
```

> **Use when**: Use this to visualize the distribution and identify potential outliers.

### Boxplot with Whiskers

```python
sns.boxplot(data=iris, x='species', y='sepal_length')
plt.show()
```

## ðŸ“Œ Special Plots

> **Use when**: Use this to show proportions of a whole, though itâ€™s not ideal for many categories.

### Pie Chart

```python
species_counts = iris['species'].value_counts()
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')

plt.show()
```

> **Use when**: Use this to identify relationships between multiple numeric variables.

### Correlation Matrix

```python
corr = iris.drop('species', axis=1).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')

plt.show()
```

> **Use when**: Use this to visualize the distribution and density of a variable split by categories.

### Violin Plot

```python
sns.violinplot(data=iris, x='species', y='sepal_length')
plt.show()
```

> **Use when**: Use this to visualize overlapping data points with random noise added for clarity.

### Jitter Plot

```python
sns.stripplot(data=iris, x='species', y='sepal_length', jitter=True)
plt.show()
```

## ðŸ“Œ Style and Layout

> **Use when**: This demonstrates how seaborn changes the default plotting style for cleaner visuals.

### Matplotlib vs. Seaborn Styles

```python
sns.reset_orig()  # Reset to matplotlib defaults
```

```python
# Matplotlib style
plt.figure()
plt.hist(iris['petal_length'], color='gray')
plt.title('Matplotlib Style')
plt.show()
```

```python
# Seaborn style
sns.set_theme()
sns.histplot(iris['petal_length'])
plt.title('Seaborn Style')
plt.show()
```

> **Use when**: Use different color palettes to improve clarity and accessibility, especially in categorical plots.

### Color Palettes

```python
sns.set_palette('Set2')
sns.boxplot(data=iris, x='species', y='petal_width', hue='species')
plt.title('Set2 Palette')
plt.show()

sns.set_palette('husl')
sns.boxplot(data=iris, x='species', y='petal_width', hue='species')
plt.title('Husl Palette')
plt.show()

sns.set_palette('Paired')
sns.boxplot(data=iris, x='species', y='petal_width', hue='species')
plt.title('Husl Palette')
plt.show()


custom_palette = ['#e41a1c', '#377eb8', '#4daf4a', '#984ea3', '#ff7f00']
sns.set_palette(custom_palette)
sns.boxplot(data=iris, x='species', y='petal_width', hue='species')
plt.title('Husl Palette')
plt.show()
```

### Subplots: Two Side-by-Side

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 4))
sns.histplot(iris['sepal_length'], ax=axs[0])
sns.histplot(iris['sepal_width'], ax=axs[1])
axs[0].set_title('Sepal Length')
axs[1].set_title('Sepal Width')
plt.tight_layout()
plt.show()
```

> **Use when**: Use this layout to display multiple plots for comparison in a grid.

### Subplots: 2x2 Grid

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
sns.boxplot(data=iris, x='species', y='sepal_length', ax=axs[0, 0])
sns.boxplot(data=iris, x='species', y='sepal_width', ax=axs[0, 1])
sns.boxplot(data=iris, x='species', y='petal_length', ax=axs[1, 0])
sns.boxplot(data=iris, x='species', y='petal_width', ax=axs[1, 1])
plt.tight_layout()
plt.show()
```

> **Use when**: Use this when you want an asymmetrical layout for better narrative flow.

### Subplots: Two Top, One Bottom

```python
fig = plt.figure(figsize=(10, 8))
ax1 = plt.subplot(2, 2, 1)
ax2 = plt.subplot(2, 2, 2)
sns.histplot(iris['sepal_length'], ax=ax1)
sns.histplot(iris['sepal_width'], ax=ax2)

ax3 = plt.subplot(2, 1, 2)
sns.histplot(iris['petal_length'], ax=ax3)
plt.tight_layout()
plt.show()
```

> **Use when**: Use facet plots to compare subsets of your data in a consistent visual framework.

### Facet Plots (Small Multiples)

```python
g = sns.FacetGrid(iris, col='species')
g.map(sns.histplot, 'sepal_length')
plt.show()
```
